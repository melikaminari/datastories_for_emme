## Και από εδώ και πέρα ξεκινούν οι δικές σας 10 ερωτήσεις στα δεδομένα! Φτιάξτε γραφήματα και βρείτε π.χ. σε ποιες βιομηχανίες συμβαίνουν τα περισσότερα ατυχήματα κλπ. Καλή επιτυχία!

1)Δείξε μου τα δεδομένα ανα στέιτ

```python
accidents.groupby('State').sum()
```

2)Ποιο επάγγελμα έχει τις περισσότερες εισαγωγές σε νοσοκομείο 

```python
accidents.plot(kind='scatter', x='Hospitalized', y='Employer')
```

3)Αν επικεντρωθούμε στο στέιτ της Νέας Υόρκης, ποια είναι τα ατυχήματα που οφείλονται σε κατάγματα?

```python
accidents[(accidents['State']== 'NewYork') & (accidents['NatureTitle']=='Fractures')].head()
```

4)Ποια είναι τα 5 στέιτ που έχουν τα περισσότερα ατυχήματα?

```python
accidents['State'].value_counts().sort_values(ascending=True).tail().plot(kind='barh', color=['green', 'pink','blue', 'magenta', 'red'])
```

5)Ποιος είναι ο συνολικός αριθμός των νοσηλειών άνα στειτ?

```python
accidents.groupby("State")["Hospitalized"].sum()
```

6)Ποιες είναι οι πιο συχνές αιτίες για ατυχήματα?


```python
accidents['SourceTitle'].value_counts().sort_values(ascending=True).plot(kind='bar')
```

7)Ποια είναι τα μέρη του σώματος που πλήττονται πιο συχνά?

```python
accidents['Part of Body Title' ].value_counts().head().plot(kind='barh', color=['yellow', 'orange','pink', 'red', 'green',])
```

8)Ποιες είναι οι 5 πόλεις που γίνονται τα περισσότερα εργατικά ατυχήματα?

```python
accidents['City'].value_counts().head().plot(kind='barh', color=['blue', 'purple', 'magenta', 'red', 'pink'])
```

9)Ποια είναι η σχέση της αιτίας του ατυχήματος με τον ακρωτηριασμό?


```python
accidents.plot(kind='scatter', x='EventTitle', y='Amputation')
```

10)Ποιες είναι οι πιο συχνές ημερομηνίες ατυχημάτων?
